```python
# comparing each network

network_classes = ["coRT", "coURL", "hashSeq", "fastRT", "tweetSim"]

for i in range(len(network_classes)):
    G = data_loader[network_classes[i]]

    print(len(G.nodes), len(G.edges), len(G.edges) * 1.0 / len(G.nodes))
```

```python
G_all = data_loader["graph"]
len(G_all.nodes), len(data_loader["labels"])
```

```python
fig, axe = plt.subplots(len(network_classes), 1, figsize=(4, 4))

for i in range(len(network_classes)):
    trace = network_classes[i]
    G = data_loader[trace].copy()
    self_edges = [(u, v) for u, v in G.edges if u == v]
    G.remove_edges_from(self_edges)
    self_nodes = [u for u in G.nodes if not G.degree()[u]]
    G.remove_nodes_from(self_nodes)
    print(
        f"Number of edges before: {len(G.edges) + len(self_edges)}, after: {len(G.edges)}"
    )
    print(
        f"Number of nodes before: {len(G.nodes) + len(self_nodes)}, after: {len(G.nodes)}"
    )

    colors = ["yellow" if data_loader['labels'][i] else "blue" for i in G.nodes]

    pos = nx.spring_layout(G)
    nx.draw(G, pos, ax=axe[i], node_size=1)
    axe[i].set_title(trace)
plt.show()
```

```python
degree = [dg for _, dg in G.degree()]

plt.hist(degree, bins=[2**i for i in range(9)])
plt.show()
```
